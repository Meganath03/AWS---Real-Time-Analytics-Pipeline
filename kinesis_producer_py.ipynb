```python
import boto3
import json
import time
from random import randint

# AWS Kinesis client setup
kinesis_client = boto3.client(
    'kinesis',
    aws_access_key_id='AKIAYWBJYP5P5GZWEQR7',
    aws_secret_access_key='HzlznPxyRp3+bU3GgM2zLEEDCGWzyZaehYNiEBJ7',
    region_name='eu-north-1'
)

# Define the Kinesis stream name
STREAM_NAME = "my-stream"  # Replace with your stream name

# Function to generate random sensor data
def generate_sensor_data():
    """Simulate random IoT sensor data."""
    return {
        "sensor_id": f"sensor-{randint(1, 100)}",
        "temperature": randint(-20, 50),
        "humidity": randint(10, 90),
        "timestamp": int(time.time())
    }

# Function to send data to Kinesis stream continuously
def send_data_to_kinesis():
    """Continuously send data to Kinesis stream."""
    try:
        while True:
            # Generate random sensor data
            data = generate_sensor_data()

            # Send the generated data to the Kinesis stream
            response = kinesis_client.put_record(
                StreamName=STREAM_NAME,
                Data=json.dumps(data),  # Convert the data to JSON string
                PartitionKey=str(data['sensor_id'])  # Use sensor_id as partition key
            )

            # Log the data sent to Kinesis and the response
            print(f"Data sent to Kinesis: {data}")
            print(f"Kinesis Response: {response}")

            # Wait for 1 second before sending the next data point
            time.sleep(1)  # Adjust the frequency of data sending

    except Exception as e:
        print(f"Error sending data to Kinesis: {e}")

# Run the function to start sending data to Kinesis
if __name__ == "__main__":
    send_data_to_kinesis()
```
