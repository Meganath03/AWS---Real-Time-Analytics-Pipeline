```python
import boto3
import json
import time
from random import randint
kinesis_client = boto3.client(
    'kinesis',
    aws_access_key_id='AKIAYWBJYP5P5GZWEQR7',
    aws_secret_access_key='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',
    region_name='eu-north-1'
)
STREAM_NAME = "my-stream"
def generate_sensor_data():
    """Simulate random IoT sensor data."""
    return {
        "sensor_id": f"sensor-{randint(1, 100)}",
        "temperature": randint(-20, 50),
        "humidity": randint(10, 90),
        "timestamp": int(time.time())
    }
def send_data_to_kinesis():
    """Continuously send data to Kinesis stream."""
    try:
        while True:
            data = generate_sensor_data()
            response = kinesis_client.put_record(
                StreamName=STREAM_NAME,
                Data=json.dumps(data),
                PartitionKey=str(data['sensor_id'])
            )
            print(f"Data sent to Kinesis: {data}")
            print(f"Kinesis Response: {response}")
            time.sleep(1)
    except Exception as e:
        print(f"Error sending data to Kinesis: {e}")
if __name__ == "__main__":
    send_data_to_kinesis()
```
